```python
import pandas as pd

df = pd.read_csv('wine.csv', header=None)

df
# 12번째 속성이 1 = 레드와인. 0 = 화이트와인
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint

import numpy as np
import os
import matplotlib.pyplot as plt
import tensorflow as tf

np.random.seed(3)
tf.random.set_seed(3)

df = df.sample(frac=0.15)

print(df.shape)
df.head()
```

---
#### DataFrame.sample( n = None , frac = None , replace = False , weights = None , random_state = None , axis = None )
---
- frac 플로트, 옵션
  반환 할 축 항목의 비율입니다. n 과 함께 사용할 수 없습니다 .

```python
# 독립변수. 종속변수 설정

X = df.values[:, 0:12]
Y = df.values[:,12]
```

```python
# 모델구조 설정, complie-fit

model = Sequential()
model.add(Dense(30, input_dim=12, activation='relu'))
# input_dim=12 ?? ==> 처음 은닉층에 들어갈 독립변수가 12개
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))
# 마지막 출력층
```

```python
# 모델 컴파일

model.compile(loss='binary_crossentropy', optimizer='adam',
             metrics=['accuracy'])
```

### 동기식 처리
---
* 동기식 처리는 말그대로 코드 한줄의 동작이 끝나면 다음 코드가 실행되는 구조입니   다.
  반면 비동기는 코드를 우선 실행시키고 결과가 나오기 전에 우선 다음 코드를 실행   합니다.
  
* 콜백함수는 비동기 처리를 구현하는 하나의 방법이라고 보시면 됩니다.
---

```python
y_value = history.history['val_loss']  # 오차값을 저장
y_acc = history.history['accuracy']    # 정확도를 저장
# 특정값만 뽑아낼 수 있다
# 이제 그래프를 그려볼 타이밍~

print(len(y_value))
print(len(y_acc))
```

```python
x_len = np.arange(len(y_acc))


plt.plot(x_len, y_value, 'o', c='red', markersize=3)
plt.plot(x_len, y_acc, 'o', c='blue', markersize=3)
```

```python
#모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

from keras.callbacks import ModelCheckpoint, EarlyStopping

#모델 저장 폴더 지정
MODEL_DIR = './model/' #현 폴더와 같은 공간에 model이라는 이름으로 저장
if not os.path.exists(MODEL_DIR) :
    os.mkdir(MODEL_DIR)
    
#모델 저장 방법
model_path = './model/{epoch:02d}-{val_loss:.4f}.hdf5'
checkpointer = ModelCheckpoint(filepath=model_path, monitor ='val_loss',verbose=1, save_best_only=True)

# 학습 조기 종료
early_stopping = EarlyStopping(monitor='val_loss',patience=100)
history = model.fit(X,Y,validation_split=0.33, epochs=1500,batch_size=500,
                    callbacks=[early_stopping, checkpointer])
history
model.summary()
```

### 모델을 불러와서 예측하고 평가하기

```python
# 모델 불러오기
model_1 = tf.keras.models.load_model('./model/16-0.0986.hdf5')

model_1.summary()
```

- from tensorflow import keras model = keras.models.load_model(‘path/to/location’)


이런식으로 import 하면서 바로 모델 불러올 수도 있어요

```python
# 예측하기

from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv('wine.csv', header=None)

np.random.seed(3)
tf.random.set_seed(3)

X = df.values[:, 0:12]
Y = df.values[:,12]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,
                                                    random_state=40)


model_1.fit(X_train, Y_train, epochs=1500, batch_size=500)
```

```python
# 테스트셋에 모델 적용

print('\n TEST ACCURACY: %.4f' % (model_1.evaluate(X_test,Y_test)[1]))
```

로직을 통해 1에 몇 프로 이상이면 레드와인 0에 몇 프로 이상이면 화이트 와인이다
