```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

df = pd.read_csv('ThoraricSurgery.csv', header = None)
train, test = train_test_split(df, test_size= 0.3)

train_arr = train.to_numpy()
test_arr = test.to_numpy()
x_train = train_arr[:, 0:17]
y_train = train_arr[:, 17]

model = Sequential()
model.add(Dense(20, input_dim = 17, activation='relu'))
model.add(Dense(1, activation= 'sigmoid'))

model.compile(loss= 'binary_crossentropy', optimizer='adam', metrics = ['accuracy'])

model.fit(x_train, y_train, epochs=100, batch_size =10)

# print('Data : ', df.shape)
# print('Train : ', train.shape)
# print('Test : ', test.shape)
# print('\n',type(train_arr))
# print(x_train)
```
