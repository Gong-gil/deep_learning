```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

df = pd.read_csv('ThoraricSurgery.csv', header = None)
train, test = train_test_split(df, test_size= 0.3)

train_arr = train.to_numpy()
test_arr = test.to_numpy()
x_train = train_arr[:, 0:17]
y_train = train_arr[:, 17]

model = Sequential()
model.add(Dense(20, input_dim = 17, activation='relu'))
# 은닉층을 하나 더 들었다.
# model.add(Dense(20, activation='softmax'))
model.add(Dense(1, activation= 'sigmoid'))

model.compile(loss= 'binary_crossentropy', optimizer='adam', metrics = ['accuracy'])

model.fit(x_train, y_train, epochs=100, batch_size =10, verbose=1)
# 'verbose = 1'이 디폴트다.

# print('Data : ', df.shape)
# print('Train : ', train.shape)
# print('Test : ', test.shape)
# print('\n',type(train_arr))
# print(x_train)
```

```python
x_test = test_arr[:, 0:17]
y_test = test_arr[:,17]

score = model.evaluate(x_test, y_test, verbose=0)
print(score)
```

```python
x_test  = test_arr[:, 0:17]

prediction = model.predict(x_test)
patient = prediction[int(input('조사할 환자의 번호를 넣으세요: '))]
patient

p = patient[0] * 100
acc = score[1] * 100

print(f'환자분의 생존률은 {p}% 입니다. 정확도는 {acc}% 입니다.')

# if prediction 
# prediction[0]
```
